```python
def clean_json(json_str):
    data = json.loads(json_str)
    data['Gender'] = data['Gender'].replace('F', "Female").replace("M", "Male")
    data['Total MMSE score'] = data['MMSE']
    del data['MMSE']
    return json.dumps()

# Apply the function to the DataFrame column
eeg["Input"] = eeg["Input"].apply(clean_json)
```
