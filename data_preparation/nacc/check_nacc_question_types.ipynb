```python
def get_question_counts(df):
    grouped = df.groupby(['Q_TYPE', 'ground_truth_text'])['ID'].nunique().reset_index(name='unique_id_count')
    all_values_counts = df.groupby(['Q_TYPE', 'ground_truth_text']).size().reset_index(name='all_values_count')
    
    np_one = df[(df['Q_TYPE'] == "Neuropath") & (~df['options'].str.contains(" and "))]
    np_mixed = df[(df['Q_TYPE'] == "Neuropath") & (df['options'].str.contains(" and "))]
    
    np_one_grouped = np_one.groupby(['Q_TYPE', 'ground_truth_text'])['ID'].nunique().reset_index(name='unique_id_count')
    np_mixed_grouped = np_mixed.groupby(['Q_TYPE', 'ground_truth_text'])['ID'].nunique().reset_index(name='unique_id_count')
    
    return grouped, all_values_counts, np_one_grouped, np_mixed_grouped
```

```python
grouped_train, all_values_counts_train, np_one_grouped_train, np_mixed_grouped_train = get_question_counts(nacc_train)
```
