```python
from pathlib import Path
import polars as pl
```

```python
medexpqa_path = Path('/projectnb/vkolagrp/projects/adrd_foundation_model/benchmarks/medexpqa/medexpqa_deduplicated.jsonl')

medexpqa = pl.read_ndjson(medexpqa_path)

num_to_letters = {
    "1": "A",
    "2": "B",
    "3": "C",
    "4": "D",
    "5": "E",
}


def make_option_string():
    expr = (
        "A. " + pl.col("A") + "\n"
        + "B. " + pl.col("B") + "\n"
        + "C. " + pl.col("C") + "\n"
        + "D. " + pl.col("D") + "\n"
        + "E. " + pl.col("E") 
    ).alias('options')

    return expr


medexpqa.select(
    "ID",
    "question",
    "options_dict",
    pl.col("ground_truth").replace_strict(num_to_letters),
).unnest("options_dict").rename(num_to_letters).with_columns(make_option_string()).select(
    'ID',
    'question',
    'options',
    'ground_truth'
).write_ndjson(medexpqa_path.parent / 'medexpqa_minimal.jsonl')
```

```python
medmcqa_path = Path('/projectnb/vkolagrp/projects/adrd_foundation_model/benchmarks/medmcqa/medmcqa_test_deduplicated.jsonl')
medmcqa = pl.read_ndjson('/projectnb/vkolagrp/projects/adrd_foundation_model/benchmarks/medmcqa/medmcqa_test_deduplicated.jsonl')
```

```python
mm = pl.read_ndjson(medmcqa_path.parent / 'dev.json')
```

```python
def make_option_string_mm():
    expr = (
        "A. " + pl.col("A") + "\n"
        + "B. " + pl.col("B") + "\n"
        + "C. " + pl.col("C") + "\n"
        + "D. " + pl.col("D") 
    ).alias('options')

    return expr

mm.rename({
    'opa':'A',
    'opb':'B',
    'opc':'C',
    'opd':'D',
}).with_columns(
    pl.col('cop').cast(str).replace( num_to_letters),
    make_option_string_mm()
).select(
    pl.col('id').alias('ID'),
    'question',
    'options',
    pl.col('cop').alias('ground_truth')
).write_ndjson(medmcqa_path.parent / 'medmcqa_minimal.jsonl')
```
